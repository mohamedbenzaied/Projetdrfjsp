```python
import numpy as np
import pandas as pd
import random
```

```python
data=pd.read_csv('submit.csv',names=['Task','Machine','Start'],header=None)
```

```python
data.head()
```

```python
df=pd.read_csv('./data/process_time_matrix.csv')
df2=pd.read_csv('./data/work_order.csv',header=None)

task=list(data['Task'])
start=list(data['Start'])
machine=list(data['Machine'])
```

```python
dur=[]
for i in range(len(task)):
    categ=df2.values[task[i]-1][2]
    mach=machine[i]
    dur.append(df.values[mach-1][categ])
```

```python
data['Finish'] = list(map(sum, zip(start, dur)))
```

```python
from datetime import datetime
from datetime import timedelta

def min_to_date(times):
    time_str = '01-01-2023 00:00:00'
    date_format_str = '%d-%m-%Y %H:%M:%S'
    given_time = datetime.strptime(time_str, date_format_str)
    print('Given timestamp: ', given_time)
    print(times)
    final_time = given_time + timedelta(minutes=times)
    print('Final Time (15 minutes after given time ): ', final_time)
    return final_time
```

```python
data['Start']=data['Start'].apply(min_to_date)
data['Finish']=data['Finish'].apply(min_to_date)

```

```python
data['Start'] = pd.to_datetime(data['Start'])
data['Finish'] = pd.to_datetime(data['Finish'])
```
